
##### Se generan p puntos de n componentes cada uno. La Regresión Lineal trata de ajustar a:




\begin{equation*}
y^p=\sum _{i=1}^pw_ix_i^p+b
\end{equation*}

#####  A partir de aquí, el término indpendiente pasa a tratarse como integrante del verctor w multiplicado siempre por una entrada constanye e igual a uno

```julia
p = 100;
n = 5;
```

#### Pesos y abcisas aleatorios

```julia
x = rand(p,n)
W = rand(n+1);
```

##### Se concatena la úkltima entrada, que es constanet e igual a uno

```julia
xe = hcat(x, ones(p,1));
y = xe*W + 0.01*rand(p);
```

#### Cálculo de las matriz del sistema de n+1 ecuaciones con n+1 incógnitas fruto del método de mínimos cuadrados aplicado al error cuadrático

El sistema de ecuaciones se puede escribir como:

\begin{equation*}
B=A\left(\begin{matrix}w_1\\w_2\\.\\.\\.\\w_n\\b\end{matrix}\right)
\end{equation*}

Donde, tras hacer las operaciones pertinentes, los términos de las matrices A y B son:

\begin{equation*}
A_{\mathit{ij}}=\sum _{v=1}^px_i^vx_j^v\text{;                         }B_i=\sum _{v=1}^py^vx_i^v
\end{equation*}
