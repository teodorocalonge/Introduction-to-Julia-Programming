#### Dado que Julia es un lenguaje pensado para Ciencia de Datos, se empezará con un programa que pinte varias funciones coseno 

```julia
using Plots
```

```julia
f(x) = 2cos.(2x.^2 .+ π/2)
g(x) = 2cos.(2x.^2 .+ π)
h(x) = 2cos.(2x.^2)
```

```julia
x = collect(range(0, π; length=1000));
```

```julia
img = plot(x, f(x), title="cos(2x^2+ z)", label="f", xlabel="x", ylabel="y")
plot!(img, x, g(x), label="g")
plot!(img, x, h(x), label="h")
#plot!(img, x, f(x), color='blue' )
```
